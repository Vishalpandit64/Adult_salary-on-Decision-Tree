```python
# Drop the row which is not having the "?"
df = df[df['workclass']!='?']
df.head()
```

```python
# Select the all columns 
df_cat = df.select_dtypes(include=['object'])
```

```python
# Check these columns have ? in the given dataset
df_cat.apply(lambda x: x=="?", axis=0).sum()
```

```python
# Remove the columns
df= df[df['occupation']!='?']
df= df[df['native.country']!='?']
```

```python
# Cleaned Data
df.info()
```

```python
# Import the libaray
from sklearn import preprocessing

# Encoding the Variable

# Select the all the columns where the categorical data is availabe;
df_cat = df.select_dtypes(include=['object'])
df_cat.head()
```

```python
# Apply the Encoding
le = preprocessing.LabelEncoder()
df_cat = df_cat.apply(le.fit_transform)
df_cat.head()
```

```python
# Concat the def_cat and Remove the columns
df = df.drop(df_cat.columns, axis=1)
df = pd.concat([df, df_cat], axis=1)
df.head()
```

```python
# Convert the Income into the category
df['income'] = df['income'].astype('category')
```

```python
df.info()
```

## Model Building and Evaluation

```python
## Let's first build a decision tree with default hyperparameters. Then we'll use cross-validation to tune them.
```

```python
from sklearn.model_selection import train_test_split
```

```python
# Putting the feature Scaling variable in X and Y
X = df.drop('income',axis=1)

y = df['income']
```

```python
# Splitting the Data in train and test
X_train , X_test ,y_train, y_test = train_test_split(X,y,test_size=.3,random_state=99)

X_train.head()
```

```python
# Import Decision Tree classifier 
from sklearn.tree import DecisionTreeClassifier
# Fitting the decision tree with default hyperparameters, apart from
# max_depth which is 5 so that we can plot and read the tree.
dt_default = DecisionTreeClassifier(max_depth=5)
dt_default.fit(X_train, y_train)
```

```python
# Import the Decision Tree 
from sklearn.metrics import classification_report , confusion_matrix, accuracy_score

# Making Predictions
y_pred_default = dt_default.predict(X_test)

# Printing classification report
print(classification_report(y_test, y_pred_default))
```

```python
# Print confusion matric and acuracy
print(confusion_matrix(y_test,y_pred_default))
print(accuracy_score(y_test,y_pred_default))
```

# Hyperparameter Grid Search CV

```python
# GridSearchCV to find optimal max_depth
from sklearn.model_selection import KFold
from sklearn.model_selection import GridSearchCV


# specify number of folds for k-fold CV
n_folds = 5

# parameters to build the model on
parameters = {'max_depth':range(1,40)}

# instantiate the model
dtree = DecisionTreeClassifier(criterion='gini', random_state=100)

# fit tree on training data
tree = GridSearchCV(dtree,parameters,cv=n_folds,scoring='accuracy',return_train_score=True)
tree.fit(X_train,y_train)
```

```python
help(sklearn.model_selection.GridSearchCV)
```

```python
# scores of GridSearch CV
scores = tree.cv_results_
pd.DataFrame(scores).head()
```

```python
# plotting accuracies with max_depth
plt.figure()
plt.plot(scores["param_max_depth"], 
         scores["mean_train_score"], 
         label="training accuracy")
plt.plot(scores["param_max_depth"], 
         scores["mean_test_score"], 
         label="test accuracy")
plt.xlabel("max_depth")
plt.ylabel("Accuracy")
plt.legend()
plt.show()
```

# Grid Search to Find Optimal Hyperparameters


```python
# Create the parameter grid 
param_grid = {
    'max_depth': range(5, 15, 5),
    'min_samples_leaf': range(50, 150, 50),
    'min_samples_split': range(50, 150, 50),
    'criterion': ["entropy", "gini"]
}

n_folds = 5

# Instantiate the grid search model
dtree = DecisionTreeClassifier()
grid_search = GridSearchCV(estimator = dtree, param_grid = param_grid, 
                          cv = n_folds, verbose = 1,return_train_score=True)

# Fit the grid search to the data
grid_search.fit(X_train,y_train)
```

```python
# cv results
cv_results = pd.DataFrame(grid_search.cv_results_)
cv_results
```

```python
# printing the optimal accuracy score and hyperparameters
print("best accuracy", grid_search.best_score_)
print(grid_search.best_estimator_)
```

```python
# model with optimal hyperparameters
clf_gini = DecisionTreeClassifier(criterion = "gini", 
                                  random_state = 100,
                                  max_depth=10, 
                                  min_samples_leaf=50,
                                  min_samples_split=50)
clf_gini.fit(X_train, y_train)
```
